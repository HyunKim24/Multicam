```python
# 연속된 날짜 리스트 획득 => 시계열 데이터 구성
import pandas as pd
```

```python
date = pd.date_range('2005-2-7',periods=2,freq='D')
date
```

```python
for d in date:
    # 우리가 원하는 포맷 YYYYmmdd
    print(d.strftime('%Y%m%d'))
```

```python
# 진행률 표시
from tqdm import tqdm_notebook
tmp = []
```

```python
for d in tqdm_notebook(date,periods=2,freq='D'):
    page = urlopen(url %d.strftime('%Y%m%d')) # 요청
    soup = BeautifulSoup(page,'html.parser')  # 수프
    divs = soup.find_all('div','tit5')        # 데이터 추출
    points = soup.find_all('td','point')      # 데이터 추출
    # 파싱
    dump = [{'time':d,
             'movie':divs[n].a.string,
             'point':points[n].string }
              for n in range(len(points))]
    tmp.extend(dump)
    
    # 확인
    # print(list[0])
    # 디비입력 (생략)
    
    
```
